```python
import json
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Create a table for the results of the samromur_asr dataset

# Load the data from the JSON file
with open('samromur_logging_step_process.json', 'r') as f:
    data = json.load(f)


# Initialize the lists for training loss, validation loss, and WER
train_loss = []
val_loss = []
wer = []
steps = []

# Extract the data
for log in data['log_history']:
    if 'loss' in log:  # This is a training step
        train_loss.append(log['loss'])
    if 'eval_loss' in log and 'eval_wer' in log:  # This is a validation step
        val_loss.append(log['eval_loss'])
        wer.append(log['eval_wer'])
        steps.append(log['step'])

# Create a DataFrame
df = pd.DataFrame({
    'Step': steps,
    'Training Loss': train_loss,
    'Validation Loss': [round(x, 4) for x in val_loss],  # round to four decimal places
    'WER': [round(x, 4) for x in wer]  # round to four decimal places
})

# Convert the 'Step' column to integers
df['Step'] = df['Step'].astype(int)

# Convert all values to strings, format decimals appropriately
df = df.applymap(lambda x: f"{x:.4f}" if isinstance(x, float) else str(x))

# Plot DataFrame as a table and remove axis
fig, ax = plt.subplots(figsize=(10, 4)) # set size frame
ax.axis('tight')
ax.axis('off')
the_table = ax.table(cellText=df.values,
         colLabels=df.columns,
         cellLoc = 'center', 
         loc='center')

# Modify the height and width of the cells in the table
table_props = the_table.properties()
table_cells = table_props['children']
for cell in table_cells:
    cell.set_height(0.10)
    cell.set_width(0.16)

# Change font size
the_table.auto_set_font_size(False)
the_table.set_fontsize(10)

plt.savefig('table.png')
```

```python


def plot_bar_chart(json_files):
    # List of labels for the bars
    labels = ['Training Loss', 'Validation Loss', 'WER']

    # Define a list of colors for the bars of each model
    colors = ['red', 'blue', 'green']

    # Initialize lists to store final values for each model
    final_training_losses = []
    final_validation_losses = []
    final_wers = []

    for json_file in json_files:
        with open(json_file, 'r') as f:
            data = json.load(f)
        
        log_history = data['log_history']

        # Extract final data
        final_training_loss = [log['loss'] for log in log_history if 'loss' in log][-1]
        final_validation_loss = [log['eval_loss'] for log in log_history if 'eval_loss' in log][-1]
        final_wer = [log['eval_wer'] for log in log_history if 'eval_wer' in log][-1]

        final_training_losses.append(final_training_loss)
        final_validation_losses.append(final_validation_loss)
        final_wers.append(final_wer)

    # Width of the bars
    width = 0.1

    # Position of bars on x-axis
    ind = np.arange(len(final_training_losses))
    
    fig = plt.figure(figsize=(10,5))
    ax = fig.add_axes([0,0,1,1])

    # Creating bars
    bar1 = [ax.bar(ind + width*i, final_values, width, color=colors[i]) for i, final_values in enumerate([final_training_losses, final_validation_losses, final_wers])]

    # Adding annotations
    ax.set_xticks(ind + width)
    ax.set_xticklabels(['malromur_asr_model', 'samromur_asr_model', 'althingi_asr_model'])
    ax.legend(labels=labels)

    # Adding labels
    plt.xlabel('Models')
    plt.ylabel('Loss/WER')
    plt.title('Comparison of Final Training Loss, Validation Loss and WER of Models')

    plt.show()

# Provide the paths to your JSON files in the list
json_files = ['malromur_logging_step_process.json', 'samromur_logging_step_process.json', 'althingi_logging_step_process.json']
plot_bar_chart(json_files)
```
